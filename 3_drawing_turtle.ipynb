```python
import turtle
import time
def draw_square(some_turtle):
    for n in range (1,5):
        some_turtle.speed(2)
        some_turtle.forward(150) 
        some_turtle.right(90)
        
def draw_circle(some_turtle):
    some_turtle.speed(2)
    some_turtle.circle(100) 
        
def draw_triangle(some_turtle):
    for n in range (1,4):
        some_turtle.left(120)
        some_turtle.speed(2)
        some_turtle.forward(300) 
        some_turtle.left(120)
    
def turtle_dance():  
    window = turtle.Screen()
    window.bgcolor("black")
    
    junnie = turtle.Turtle()
    junnie.shape("turtle")    
    junnie.color("pink")
    
    derek = turtle.Turtle()
    derek.shape("turtle")    
    derek.color("teal")
        
    bo = turtle.Turtle()
    bo.shape("turtle")    
    bo.color("orange")

    draw_square(junnie)
    draw_circle(derek)
    draw_triangle(bo)
    
    window.exitonclick()
    
turtle_dance()
    
```

```python
import turtle
import time
def draw_square(some_turtle):
    for n in range (1,5):
        some_turtle.speed(5)
        some_turtle.forward(150) 
        some_turtle.right(90)

def draw_circle(some_turtle):
    some_turtle.speed(5)
    some_turtle.circle(100) 
        
def draw_triangle(some_turtle):
    for n in range (1,4):
        some_turtle.left(120)
        some_turtle.speed(5)
        some_turtle.forward(300) 
        some_turtle.left(120)
        
def turtle_dance(s,c,t):  
    window = turtle.Screen()
    window.bgcolor("white")    
    
    draw_square(s)
    draw_circle(c)
    draw_triangle(t)

    window.exitonclick()
    
junnie = turtle.Turtle()
junnie.shape("turtle")    
junnie.color("pink")
    
derek = turtle.Turtle()
derek.shape("turtle")    
derek.color("teal")
        
bo = turtle.Turtle()
bo.shape("turtle")    
bo.color("orange")

# turtle_dance(square, circle, triangle)
turtle_dance(derek, junnie, bo)
    
```
